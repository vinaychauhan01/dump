<h3>Deploy Rex Encoder Bot on Heroku (Multi App Support)</h3>

```python
#@title <center><h3>Heroku Login</h3></center><br>
Heroku_Email = "" #@param {type:"string"}
Heroku_API = "" #@param {type:"string"}

!curl -s https://cli-assets.heroku.com/install.sh | sh
from IPython.display import HTML, clear_output; clear_output()
display(HTML("<h3>Heroku CLI Installed</h3>"))

if not Heroku_Email or not Heroku_API:
    raise Exception("Heroku Email & API Key are mandatory!")

from os import path as ospath, chmod
netrc = "machine api.heroku.com\n  login {}\n  password {}\n".format(Heroku_Email, Heroku_API)
with open(ospath.expanduser("~/.netrc"), "w") as f:
    f.write(netrc*2)
chmod(ospath.expanduser("~/.netrc"), 0o600)

!git config --global user.email {Heroku_Email}
!git config --global user.name "RexBot"
display(HTML("<marquee><b>Heroku Login Successful!</b></marquee>"))
```

```python
#@title <center><h3>Create Heroku Apps (Multi)</h3></center><br>
App_Names = "" #@param {type:"string"}
Server_Region = "eu" #@param ["eu", "us"]
Team_Name = "" #@param {type:"string"}

team = f"--team {Team_Name}" if Team_Name else ""
for app in App_Names.split():
    !heroku create --region $Server_Region --stack container $team {app.strip()}
```

```python
#@title <center><h3>Rex Bots Config Setup (3 Methods)</h3></center><br>
App_Name = "" #@param {type:"string"}

#@markdown ### Method 1: Fill Manually (Recommended)
API_ID = 12345678 #@param {type:"integer"}
API_HASH = "" #@param {type:"string"}
BOT_TOKEN = "" #@param {type:"string"}
OWNER_ID = 1234567890 #@param {type:"integer"}
SUDO_USERS = "" #@param {type:"string"}
EVERYONE_CHATS = "" #@param {type:"string"}
LOG_CHANNEL = -1001234567890 #@param {type:"integer"}
SESSION_NAME = "encoderbot" #@param {type:"string"}
MONGO_URI = "" #@param {type:"string"}
DOWNLOAD_DIR = "VideoEncoder/downloads/" #@param {type:"string"}
ENCODE_DIR = "VideoEncoder/encodes/" #@param {type:"string"}
INDEX_URL = "" #@param {type:"string"}
DRIVE_DIR = "" #@param {type:"string"}

#@markdown ---
#@markdown ### Method 2: From Private Gist URL (config.env)
Gist_URL = "" #@param {type:"string"}

#@markdown ---
#@markdown ### Method 3: Upload config.env File
Upload_Config = False #@param {type:"boolean"}

#@markdown ---

import os, shutil
if os.path.exists(App_Name):
    shutil.rmtree(App_Name)

!git clone https://github.com/SilentDemonSD/Rex-Bots $App_Name
%cd $App_Name

# Clean unwanted files
for file in ["README.md", "wzv3_hk_deploy.ipynb", "deploy.ipynb"]:
    if os.path.isfile(file):
        os.remove(file)

if Upload_Config:
    from google.colab import files
    uploaded = files.upload()
    for fn in uploaded.keys():
        os.rename(fn, "config.env")
    print("config.env uploaded successfully!")

elif Gist_URL.strip():
    !curl -L -o config.env "$Gist_URL"
    print("config.env downloaded from Gist!")

else:
    # Manual config creation
    config_text = f'''# Basics #
API_ID = {API_ID}
API_HASH = "{API_HASH}"
BOT_TOKEN = "{BOT_TOKEN}"
# Authorization #
OWNER_ID = {OWNER_ID}
SUDO_USERS = {SUDO_USERS}
EVERYONE_CHATS = {EVERYONE_CHATS}
# Log Channel #
LOG_CHANNEL = {LOG_CHANNEL}
# Database #
SESSION_NAME = "{SESSION_NAME}"
MONGO_URI = "{MONGO_URI}"
# Optional #
DOWNLOAD_DIR = "{DOWNLOAD_DIR}"
ENCODE_DIR = "{ENCODE_DIR}"
INDEX_URL = "{INDEX_URL}"
DRIVE_DIR = "{DRIVE_DIR}"
'''
    with open("config.env", "w") as f:
        f.write(config_text)
    print("config.env created from manual input!")

%cd ..
print("\nAll Configured Apps:")
!ls -d */
```

```python
#@title <center><h3>Deploy to Heroku (Multi Apps)</h3></center><br>
Deploy_Apps = "" #@param {type:"string"}

for app in Deploy_Apps.split():
    app = app.strip()
    if os.path.isdir(app):
        %cd {app}
        !git add . -f
        !git commit -m "Rex Bots Deploy"
        !heroku git:remote -a {app}
        !git push heroku main -f
        print(f"{app} deployed successfully!")
        %cd ..
    else:
        print(f"Folder {app} not found! Run config cell first.")
```
